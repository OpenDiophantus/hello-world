```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout

# Assuming the number of unique classes (Chinese characters)
num_classes = 15  # Adjust this according to your actual number of Chinese characters

model = Sequential([
    # Flatten layer to convert the 64x64 images into a 1D array of 4096 pixels
    Flatten(input_shape=(4096,)),

    # First dense layer with ReLU activation
    Dense(512, activation='relu'),
    Dropout(0.5),  # Dropout layer to prevent overfitting

    # Second dense layer with ReLU activation
    Dense(256, activation='relu'),
    Dropout(0.5),  # Another dropout layer to prevent overfitting

    # Output layer with softmax activation for multi-class classification
    Dense(num_classes, activation='softmax')
])

# Compile the model with the sparse_categorical_crossentropy loss function
model.compile(optimizer='adam',                  # Optimizer
              loss='sparse_categorical_crossentropy',  # Loss function for integer labels
              metrics=['accuracy'])                     # Metric to monitor

# Print model summary to check the architecture
model.summary()
```

```python
# Valid label range from 0 to 14
valid_labels = np.arange(15)  # Creates an array [0, 1, 2, ..., 14]

# Create masks for valid labels
valid_train_mask = np.isin(y_train, valid_labels)
valid_test_mask = np.isin(y_test, valid_labels)

# Apply masks
X_train_filtered = X_train[valid_train_mask]
y_train_filtered = y_train[valid_train_mask]
X_test_filtered = X_test[valid_test_mask]
y_test_filtered = y_test[valid_test_mask]

# Check the shapes of the filtered datasets
print("Filtered X_train shape:", X_train_filtered.shape)
print("Filtered y_train shape:", y_train_filtered.shape)
print("Filtered X_test shape:", X_test_filtered.shape)
print("Filtered y_test shape:", y_test_filtered.shape)
# Assuming the model architecture is already defined, recompile it
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Retrain the model with the cleaned and filtered data
history = model.fit(
    X_train_filtered, y_train_filtered,
    epochs=25,
    batch_size=64,
    validation_data=(X_test_filtered, y_test_filtered),
    verbose=1
)
```

```python
import matplotlib.pyplot as plt
# Reset index of the filtered dataframes to ensure continuous indexing
X_test_filtered.reset_index(drop=True, inplace=True)
y_test_filtered.reset_index(drop=True, inplace=True)

# Assuming history contains the results from the model.fit() process
# and the training process has successfully completed.

# Plot loss curve for MLP
plt.figure(figsize=[8, 4])  # Set the figure size for better readability

# Loss Curves
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot for Loss
plt.plot(history.history['loss'], 'black', linewidth=2.0)
plt.plot(history.history['val_loss'], 'green', linewidth=2.0)
plt.title('Loss Curves', fontsize=14)
plt.xlabel('Epochs ', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.legend(['Training Loss', 'Validation Loss'], fontsize=10)

# Accuracy Curves
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot for Accuracy
plt.plot(history.history['accuracy'], 'black', linewidth=2.0)
plt.plot(history.history['val_accuracy'], 'blue', linewidth=2.0)
plt.title('Accuracy Curves', fontsize=14)
plt.xlabel('Epochs ', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.legend(['Training Accuracy', 'Validation Accuracy'], fontsize=10)

plt.tight_layout()  # Adjust layout to make room for titles/labels
plt.show()
```
